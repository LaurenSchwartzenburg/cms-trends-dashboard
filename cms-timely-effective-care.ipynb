```python
import pandas as pd
```

```python
# Read and display top 5 rows of csv data
df = pd.read_csv("Timely_and_Effective_Care-Hospital.csv", low_memory=False)
df.head()
```

```python
# Drop excess columns
df.drop(columns=['Measure ID', 'Footnote'], inplace=True)
```

```python
# Filter DataFrame to only include the 4 key health conditions for dashboard use
four_conditions = [
    'Cataract surgery outcome',
    'Colonoscopy care',
    'Emergency Department',
    'Healthcare Personnel Vaccination'
]

df = df[df['Condition'].isin(four_conditions)]
```

```python
# Export cleaned data to new CSV for Excel/dashboard review
df.to_csv("Timely_and_Effective_Care-Hospital.csv", index=False)
```
